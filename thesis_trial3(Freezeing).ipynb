```python
def clean_text(s: str) -> str:
    s = str(s)
    s = re.sub(r"\s+", " ", s).strip()          # normalize whitespace
    s = s.replace("\u00a0", " ")                # non-breaking spaces
    return s

df_hl["abstract_clean"] = df_hl["abstract"].apply(clean_text)

# drop empty or invalid abstracts
df_hl = df_hl[df_hl["abstract_clean"].str.len() > 20].copy()
df_hl.shape
```

```python
# binary target: accept=1, reject=0
label_map = {"accept": 1, "reject": 0}
df_hl = df_hl[df_hl["label_accept_reject"].isin(label_map.keys())].copy()
df_hl["y"] = df_hl["label_accept_reject"].map(label_map).astype(int)

df_hl["y"].value_counts()
```

```python
train_df, test_df = train_test_split(
    df_hl,
    test_size=0.2,
    random_state=SEED,
    stratify=df_hl["y"]
)

train_df, val_df = train_test_split(
    train_df,
    test_size=0.15,
    random_state=SEED,
    stratify=train_df["y"]
)

print("Train:", train_df.shape)
print("Val:", val_df.shape)
print("Test:", test_df.shape)
```

```python
import numpy as np

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=1)
    return {
        "accuracy": metric_acc.compute(predictions=preds, references=labels)["accuracy"],
        "f1": metric_f1.compute(predictions=preds, references=labels, average="binary")["f1"],
        "precision": metric_prec.compute(predictions=preds, references=labels, average="binary")["precision"],
        "recall": metric_rec.compute(predictions=preds, references=labels, average="binary")["recall"],
    }
```

# Build model + Freeze first 4 layers (top 2 + head trainable)

**Load model and freeze first 4 transformer layers**

```python
from transformers import AutoModelForSequenceClassification

MODEL_NAME = "climatebert/distilroberta-base-climate-f"
model = AutoModelForSequenceClassification.from_pretrained(MODEL_NAME, num_labels=2)

def freeze_first_n_transformer_layers(model, n_freeze=4):
    backbone = getattr(model, model.base_model_prefix)

    # DistilRoBERTa / DistilBERT style
    if hasattr(backbone, "transformer") and hasattr(backbone.transformer, "layer"):
        layers = backbone.transformer.layer
        for i in range(min(n_freeze, len(layers))):
            for p in layers[i].parameters():
                p.requires_grad = False
        print(f"Detected Distil-style. Total layers={len(layers)}. Froze first {n_freeze}.")
        return len(layers)

    # RoBERTa/BERT style
    if hasattr(backbone, "encoder") and hasattr(backbone.encoder, "layer"):
        layers = backbone.encoder.layer
        for i in range(min(n_freeze, len(layers))):
            for p in layers[i].parameters():
                p.requires_grad = False
        print(f"Detected BERT/RoBERTa-style. Total layers={len(layers)}. Froze first {n_freeze}.")
        return len(layers)

    raise ValueError("Unknown backbone structure")

n_total = freeze_first_n_transformer_layers(model, n_freeze=4)

# sanity: show how many parameters remain trainable
total_params = sum(p.numel() for p in model.parameters())
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"Trainable params: {trainable_params:,} / {total_params:,}")
```

**Custom Trainer: weighted CE + layerwise LR**

```python
import torch
import torch.nn as nn
from transformers import Trainer

class WeightedLayerwiseLRTrainer(Trainer):
    def __init__(self, class_weights, base_lr=1e-5, head_lr=2e-4, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.class_weights = class_weights
        self.base_lr = base_lr
        self.head_lr = head_lr

    def compute_loss(self, model, inputs, return_outputs=False, **kwargs):
        labels = inputs.get("labels")
        outputs = model(**inputs)
        logits = outputs.get("logits")
        loss_fct = nn.CrossEntropyLoss(weight=self.class_weights.to(logits.device))
        loss = loss_fct(logits.view(-1, model.config.num_labels), labels.view(-1))
        return (loss, outputs) if return_outputs else loss

    def create_optimizer(self):
        if self.optimizer is None:
            backbone = getattr(self.model, self.model.base_model_prefix)

            base_params = [p for p in backbone.parameters() if p.requires_grad]
            head_params = []
            for name, p in self.model.named_parameters():
                if not name.startswith(self.model.base_model_prefix) and p.requires_grad:
                    head_params.append(p)

            optimizer_grouped_parameters = [
                {"params": base_params, "lr": self.base_lr},
                {"params": head_params, "lr": self.head_lr},
            ]

            self.optimizer = torch.optim.AdamW(optimizer_grouped_parameters)
        return self.optimizer
```

```python
import numpy as np
import torch
from sklearn.metrics import f1_score, precision_score, recall_score

def get_probs_labels(trainer, dataset):
    out = trainer.predict(dataset)
    probs = torch.softmax(torch.tensor(out.predictions), dim=1).numpy()[:, 1]  # P(Accept)
    labels = out.label_ids
    return probs, labels

val_probs, val_labels = get_probs_labels(trainer, val_ds)

thresholds = np.linspace(0.05, 0.95, 91)
best = {"thr": 0.5, "f1": -1, "precision": 0, "recall": 0}

for thr in thresholds:
    pred = (val_probs >= thr).astype(int)
    f1 = f1_score(val_labels, pred, pos_label=1, zero_division=0)
    if f1 > best["f1"]:
        best = {
            "thr": float(thr),
            "f1": float(f1),
            "precision": float(precision_score(val_labels, pred, pos_label=1, zero_division=0)),
            "recall": float(recall_score(val_labels, pred, pos_label=1, zero_division=0)),
        }

best
```

```python
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

test_probs, test_labels = get_probs_labels(trainer, test_ds)
thr = best["thr"]
test_pred = (test_probs >= thr).astype(int)

print(f"Best threshold from VAL: {thr:.2f}")
print("VAL metrics @ best thr:", best)
print("\nTEST report:")
print(classification_report(test_labels, test_pred, target_names=["Reject","Accept"], zero_division=0))

cm = confusion_matrix(test_labels, test_pred)
ConfusionMatrixDisplay(cm, display_labels=["Reject","Accept"]).plot()
plt.title("Test Confusion Matrix (Thresholded)")
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from transformers import Trainer

class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=2.0):
        super().__init__()
        self.alpha = alpha
        self.gamma = gamma

    def forward(self, logits, targets):
        ce = F.cross_entropy(logits, targets, reduction="none", weight=self.alpha)
        pt = torch.exp(-ce)
        loss = ((1 - pt) ** self.gamma) * ce
        return loss.mean()

class FocalTrainer(Trainer):
    def __init__(self, alpha=None, gamma=2.0, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.focal = FocalLoss(alpha=alpha, gamma=gamma)

    def compute_loss(self, model, inputs, return_outputs=False, **kwargs):
        labels = inputs.get("labels")
        outputs = model(**inputs)
        logits = outputs.get("logits")
        loss = self.focal(logits, labels)
        return (loss, outputs) if return_outputs else loss
```

```python
from transformers import AutoModelForSequenceClassification
import torch

model = AutoModelForSequenceClassification.from_pretrained(MODEL_NAME, num_labels=2)

# Move class_weights to the appropriate device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
class_weights_on_device = class_weights.to(device)

trainer = FocalTrainer(
    model=model,
    args=training_args,
    train_dataset=train_ds,
    eval_dataset=val_ds,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
    alpha=class_weights_on_device,   # Use the weights on the correct device
    gamma=2.0
)

trainer.train()
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
import numpy as np

def threshold_for_recall(val_probs, val_labels, recall_target=0.70):
    thresholds = np.linspace(0.05, 0.95, 91)
    candidates = []
    for thr in thresholds:
        pred = (val_probs >= thr).astype(int)
        r = recall_score(val_labels, pred, pos_label=1, zero_division=0)
        p = precision_score(val_labels, pred, pos_label=1, zero_division=0)
        f = f1_score(val_labels, pred, pos_label=1, zero_division=0)
        if r >= recall_target:
            candidates.append((thr, f, p, r))
    # pick highest precision among those meeting recall target
    if not candidates:
        return None
    thr, f, p, r = sorted(candidates, key=lambda x: (x[2], x[1]), reverse=True)[0]
    return {"thr": float(thr), "f1": float(f), "precision": float(p), "recall": float(r)}

recall_choice = threshold_for_recall(val_probs, val_labels, recall_target=0.70)
recall_choice
```
